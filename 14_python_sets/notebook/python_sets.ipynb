# Sets

In Python, a set is an unordered collection of unique elements. It is defined by enclosing comma-separated values within curly braces {} or by using the set() constructor. Sets are mutable, meaning you can add or remove elements from them. However, the elements themselves must be immutable (e.g., numbers, strings, or tuples).

#### a set is an unordered collection
    Sets do  not maintain the order in which elements are inserted, as a result, sets are not suitable for indexing and slicing operations.

#### unique elements.
    sets do not allow duplicate values

#### creating a set

```python
s = {'Puttu','Kadala','Dosa',234,4}
print(s)
type(s)
```

```python
# using set constructor (from iterables such as list,tuple or string)
s1 = set([12,34,56,23,23])
print(s1)

s2 = set((12,3445,65,23,12))
print(s2)

s3 = set('Akshay Pattayil')
print(s3)
```

```python
# creating an empty set

s  = set()
print(s,type(s))
```

```python
k = {}
type(k)
```

#### Adding elements to a set

```python
s = {'Puttu','Kadala','Dosa',234,4}
print(s)
```

```python
s.add('Chammanthi')
s
```

#### Removing elements from a set

```python
s.remove(234)
s
```

```python
s.discard(4)
s
```

#### the elements themselves must be immutable (e.g., numbers, strings, or tuples).

```python
s1 = {2334,'asddf',(12,34,'df')}
s1
```

```python
s2 = {232,12,[34,231,'df']}
s2
```
