## Variables

In Python, a variable is a named reference to a value that can be assigned or changed. Variables are like containers that can store different types of data such as numbers, strings, and objects.

- Python variables can be used to store any type of value and can be referenced by their name 
- Unlike other languages such as C, Java - Variables in Python don't need initialization.
- variables don't need to be initialized before they are used. This is because Python is a dynamically typed language, which means that the type of a variable is determined at runtime based on the value assigned to it

## Data types

data types are the classifications of different types of data that can be used in a program. They define the kind of value that a variable can hold, and the operations that can be performed on that value. 

##### Common data types in python
- `Integers (int)` - whole numbers with no decimal point, such as 1, 2, 3, -4, -5, etc.
- `Floats (float)` - numbers with decimal points, such as 3.14159, 2.5, -0.1, etc.
- `Strings (Str)` - sequences of characters, such as "Hello, world!", "123", "Python", etc.
- `Booleans (bool)` - values that represent truth or falsehood, such as True and False.

##### Type casting in python

Type casting, also known as type conversion, is the process of converting one data type to another data type in a programming language. In Python, you can use built-in functions to convert a value from one data type to another. Here are some common type casting functions in Python:

- int() - converts a value to an integer data type. For example, int(3.14) will return 3.
- float() - converts a value to a floating-point data type. For example, float(3) will return 3.0.
- str() - converts a value to a string data type. For example, str(42) will return "42".
- bool() - converts a value to a boolean data type. For example, bool(0) will return False.

```python
a = 1
print(type(a))
print ( bool(a))
print ( float(a))
print ( str(a))
```
