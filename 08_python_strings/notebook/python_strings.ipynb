#### Indexing

In Python, indexing is a way to access individual elements of a sequence, such as a string. To index a string in Python, you can use square brackets [] after the string, with the index of the character you want to access inside the brackets.

- In Python, indexing starts at 0, so the first character has an index of 0, the second character has an index of 1, and so on.

```python
st = 'Akshay Pattayil'
```

```python
# access the first character in the string,
st[0]
```

```python
# Acccess the seventh character
st[6]
```

```python
# Access the 8th character
st[7]
```

- You can also use negative indices (reverse indexing) to index a string from the end. 
- Reverse indexing starts with the last character as -1 in a python string

```python
# Acess the last character
st[-1]
```

```python
#get P
st[-8]
```

```python
st[-9]
```

#### Strings are immutable

In Python, strings are immutable, which means that once a string object is created, its contents cannot be modified. This means that you cannot change individual characters in a string, add new characters to a string, or remove characters from a string.

```python
st
```

```python
# this will raise an eroor
st[3] = 'p'
```

This is because strings are immutable, and their individual elements cannot be changed.

```python
new_st = st.replace('h','p')
new_st
```

Immutable strings have some advantages, such as making the string data type more memory-efficient and thread-safe.

#### Slicing in strings

In Python, slicing is a way to extract a portion of a string (or any other sequence) by specifying a range of indices. Slicing allows you to extract a substring from a larger string, or to extract a subsequence from a larger sequence.

- Specify the start index and the end index, separated by a colon, to return a part of the string.

```python
string = 'If it was easy everyone would have done it'
```

```python
# extract 'easy' from string
string[10:14]
```

```python
# extract 'it' from string
string[3:5]
```

```python
# extract 'If it was easy'
print(string[0:14])
print(string[:14])
```

```python
# Extract 'would have done it'
print(string[24:42])
print(string[24:])
```

```python
# extract 'was easy' from string and add 'Paper' infront of it
'Paper ' + string[6:14]
```

#### Membership Operator

In Python, you can check if a substring is present in a string using the membership operator in. The in operator returns a boolean value of True if the substring is found in the string, and False otherwise.

```python
string = 'shukoor sundarana oan oru vallatha sambhavama'
```

```python
# check if the string contains the ord 'sambhavam'
'sambhavam' in string
```

```python
# check if 'Akshay' in string
'Akshay' in string
```

```python
string = 'akshay Pattayil pwolikkum'
```

Python also provides the join() method to concatenate a list of strings into a single string, using a separator string as a delimiter.

```python
my_list = ['apple', 'banana', 'cherry']
separator = ', '
result = separator.join(my_list)
print(result)  # Output: 'apple, banana, cherry'
```
