- Shallow copy is a method of creating a new object that references the same elements as the original object. It creates a new container but retains the references to the original elements, rather than duplicating them. As a result, changes made to the elements of the new object may affect the original object, and vice versa.
- A shallow copy creates a new list object, but the elements inside the list are still references to the same objects as in the original list.

# Deep copy

it creates an entirely independent copy where all the elements and their nested objects are duplicated, rather than just copying references. Deep copy ensures that modifications made to the copied object do not affect the original object or any of its nested objects.

```python
l1 = [2,[3,4],5]
```

```python
import copy
l2 = copy.deepcopy(l1)
```

```python
l2[-1] =1000
l2
```

```python
l1
```

```python
l2[1][-1] = 50
```

```python
l2
```

```python
l1
```

### Shallow Copy:

- Creates a new object that references the same elements as the original object.
- Modifying the elements in the copied object may affect the original object, and vice versa.
- It copies the references to nested objects rather than duplicating the nested objects themselves.
- Shallow copy can be performed using methods like slicing ([:]), the copy() method for lists, or specific constructors.
- Shallow copy is relatively faster and more memory-efficient than deep copy.

### Deep Copy:

Creates a new object and recursively duplicates all nested objects within it.
The copied object is entirely independent of the original object, and modifications to one do not affect the other.
It creates new instances of nested objects rather than copying their references.
Deep copy can be performed using the deepcopy() function from the copy module.
Deep copy is slower and consumes more memory compared to shallow copy, as it creates a completely separate copy of the objects.
