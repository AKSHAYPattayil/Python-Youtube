#### order of operations

order in which arithmetic operations should be performed  : PEDMAS
- The acronym stands for Parentheses, Exponents, Multiplication and Division (from left to right), and Addition and Subtraction (from left to right).
- In Python, you can use parentheses to explicitly specify the order in which arithmetic operations should be performed.

```python
c = (4+6)**2/3-8*2
```
