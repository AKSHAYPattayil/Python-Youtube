 <h1 style = 'color:red'> Basic Properties of Python
   

#### 1. Readability: Python code is designed to be easy to read and understand, with a syntax that is simple and consistent.

#### 2. Dynamic typing:  Python is a dynamically typed language, which means that the data type of a variable is determined at runtime, rather than being declared explicitly.

#### 3. Interpreted: Python is an interpreted language, which means that code is executed directly by an interpreter without the need for compilation.

When a programming language is compiled, the source code is transformed into machine code that can be executed by the computer's processor. Compiled languages like C and C++ require a separate compilation step before the code can be run.

On the other hand, interpreted languages like Python do not require a separate compilation step. Instead, the source code is interpreted directly by an interpreter program. The interpreter reads the code line by line and executes it in real-time.

#### 4. Python language runs line by line, so incase you want to update a piece of code all you have to do is update that line

```python
print('*****')
print('*$*' )
print('**')
```

#### 5. Object-oriented: Python is an object-oriented language, which means that it uses objects to represent and manipulate data.

#### 6. Portable: Python code can be run on different platforms, including Windows, macOS, and Linux, without the need for changes to the code.

#### 7. Large standard library: Python has a large standard library that provides a wide range of built-in functions and modules for performing various tasks, such as file I/O, networking, and web development.
